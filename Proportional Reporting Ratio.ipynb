Suppose we are investigating whether different drugs cause different side effects. For each drug $X$ we have reported cases of side effect $A$ $R(A \mid X)$; the reported cases of A for all the drugs which are not $X$, $R(A \mid \neg X)$; all reports of X, $R(X)$; and all reports for all drugs which are not $X$, $R(\neg X)$. 

Then, the Proportional Reporting Ratio (PRR) is given by:

$$
\text{PRR} = \frac{R(A \mid X)}{R(X)} \div \frac{R(A \mid \neg X)}{R(\neg X)}
$$

A PRR of greater than 1 is potentially evidence that $X$ is associalted with $A$. This does not establish causality (perhaps doctors and patients believe $X$ causes $A$ and so are more likely to report this) and may not necessarily be statistically significant. To establish statistical significance we can look at Chi Square (3.84 gives a p-value < 0.05) or the number of samples (we need at least 21 in each cteunder Central Limit Theorem.

```python
def calculate_proportional_reporting_ratio(
    reports_A_given_X,
    reports_A_given_notX,
    reports_X,
    reports_notX
):
    fraction_1 = reports_A_given_X / reports_X
    fraction_2 = reports_A_given_notX / reports_notX
    return(fraction_1 / fraction_2)

result = calculate_proportional_reporting_ratio(
    25,  #R(A|X)
    800, #R(A|!X)
    120, #R(X)
    8000 #R(!X)
)
print(f"Calculated PRR is: {result:.2f}")
```
