Suppose we are comparing reports of a drug $X$ and a side effect $A$, we can define:

$a = X \land A$

$b = X \land \neg A$

$c = \neg X \land A$

$d = \neg X \land \neg A$

Then our reporting odds ratio (ROR) is given by:

$ROR = \frac{a}{b} \div \frac{c}{d}$

A ROR of greater than 1 is potentially evidence that $X$ is associalted with $A$. This does not establish causality (perhaps doctors and patients believe $X$ causes $A$ and so are more likely to report this) and may not necessarily be statistically significant. To establish statistical significance we can look at Chi Square (3.84 gives a p-value < 0.05) or the number of samples (we need at least 21 in each category for statistical significance under Central Limit Theorem).

```python
def calculate_reporting_odds_ratio(a, b, c, d):
    return((a/b) / (c/d))

result = calculate_reporting_odds_ratio(
    20,
    180,
    40,
    400
)
print(f"Calculated ROR is: {result:.2f}")
```
